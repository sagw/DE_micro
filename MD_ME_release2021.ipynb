```r
sums=as.data.frame(aggregate(Count~beetag+Condition+Source, data=cells, FUN=sum))

```

```r
sums
```

```r
sums$Temperature=ifelse(sums$Condition=="ambient", "23", "28")
```

```r
sums$Condition=NULL
```

```r
colnames(sums)=c("ID", "Source", "number", "Temperature")
```

```r
sums=sums[,c(1,2,4,3)]
```

```r
all=rbind(sums, maxcells)
```

```r
all$perhour=all$number/6
```

```r
all$logcells=log10(all$number)
```

```r
all
```

```r
all$Temperature=as.numeric(all$Temperature)
```

```r
ggplot(all, aes(x = Temperature, y = perhour, color=Source)) +
scale_fill_manual(values=c("#63719A", "#C5A5B3"))+
geom_point(stat="identity", position="dodge")+
theme_bw()+
    geom_smooth(se = FALSE, method = "loess", span=10, formula = y ~ x)+
facet_grid( . ~ Source, scales="free", space="free",)+
 theme(legend.position="left",  
          panel.grid.major.x = element_blank(), 
          panel.grid.minor.x = element_blank(),
          panel.border=element_rect(colour="darkgrey",size=1),
          strip.background = element_blank(),
          axis.ticks.x=element_blank(), axis.line=element_line(color="black"), axis.text.x=element_blank(),
       text = element_text(size=20))+  
ylab("Number of cells released per hour") 
```

```r
ggplot(all, aes(x = Temperature, y = logcells, color=Source)) +
scale_color_manual(values=c("darkred", "cornflowerblue"))+
geom_point(stat="identity", position="dodge", size=3)+
theme_bw()+
scale_x_continuous(breaks=c(23,25, 28, 30))+

    geom_smooth(se = TRUE, method = "loess", span=200, formula = y ~ x)+
#facet_grid( . ~ Source, scales="free", space="free",)+
 theme(legend.position="top",  
          panel.grid.major.x = element_blank(), 
          panel.grid.minor.x = element_blank(),
          panel.border=element_rect(colour="darkgrey",size=1),
          strip.background = element_blank(),
           axis.line=element_line(color="black"), axis.text.x=element_text(), 
       text = element_text(size=25))+  
ylab("Log10 cells released") 
```

```r
sums$Source=as.factor(sums$Source)
```

```r

Tukey = contrMat(table(sums$Condition), "Tukey")
K1 = cbind(Tukey, matrix(0, nrow = nrow(Tukey), ncol = ncol(Tukey)))
rownames(K1) = paste(levels(sums$Source)[1], rownames(K1), sep = ":")
K2 = cbind(matrix(0, nrow = nrow(Tukey), ncol = ncol(Tukey)), Tukey)
rownames(K2) = paste(levels(sums$Source)[2], rownames(K2), sep = ":")
K = rbind(K1, K2)
colnames(K) = c(colnames(Tukey), colnames(Tukey))
```
